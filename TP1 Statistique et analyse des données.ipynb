```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster
```

```python
X, label = make_blobs(n_samples=100, centers=2, cluster_std=.5, n_features=2,
 random_state=0) 
```

```python
X
```

# La méthode linkage permet de générer, selon la méthode choisie single ( saut min) ou complete ( saut max), une matrice contenant les différents paires de clusters avec leur distances euclidiennes ou mahalanobis selon la métrique utilisée.

```python
Z1=linkage(X, method='single', metric='euclidean')
Z2=linkage(X, method='complete', metric='euclidean')
Z3=linkage(X, method='single', metric='mahalanobis')
Z4=linkage(X, method='complete', metric='mahalanobis')
#method c'est pour définir le saut min ou max
#metric correspond à la méthode pour le calcul de la distance 
```

# Ci-aprés les différents dendogrammes selon les deux méthodes choisies ainsi que les métriques utilisées

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram single/euclidean')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
 Z1,
 leaf_rotation=90., # rotates the x axis labels
 leaf_font_size=8. # font size for the x axis labels
   
    
)
plt.show()
```

# Le regroupement hiérarchique peut être représenté par un dendrogramme. 

# Le dendrogramme illustre la composition de chaque cluster en traçant un lien en forme de U entre un cluster non-singleton et ses enfants. Le haut du lien U indique une fusion de cluster. Les deux branches du lien en U indiquent quels clusters ont été fusionnés. La longueur des deux branches du lien en U représente la distance entre les clusters enfants. 

# Python nous propose par des jeux de couleurs une proposition de découpage en clusters
# Couper un dendrogramme à un certain niveau donne un ensemble de clusters. Couper à un autre niveau donne un autre ensemble de clusters. Comment choisir où couper le dendrogramme ?

# Néanmoins, on peut utiliser les paramétres  truncate_mode = 'level' et  p = 3 du dendogramme pour le tronquer à un niveau donné.

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram complete/euclidean')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
 Z2,
 leaf_rotation=90., # rotates the x axis labels
 leaf_font_size=8., # font size for the x axis labels
)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram single/mahalanobis')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
 Z3,
 leaf_rotation=90., # rotates the x axis labels
 leaf_font_size=8., # font size for the x axis labels
)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram complete/mahalanobis')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
 Z4,
 leaf_rotation=90., # rotates the x axis labels
 leaf_font_size=8., # font size for the x axis labels
)
plt.show()
```

# Ci-aprés la représentation des clusters pour les différents linkages utilisés en prenant comme critère la distance entre les clusters

```python
max_d = 50
clusters01 = fcluster(Z1, max_d, criterion='distance')
max_d = 50
clusters02 = fcluster(Z2, max_d, criterion='distance')
max_d = 50
clusters03 = fcluster(Z3, max_d, criterion='distance')
max_d = 50
clusters04 = fcluster(Z4, max_d, criterion='distance')
```

```python
plt.scatter(X[:,0], X[:,1], c=clusters01, cmap='prism')  
plt.show()
plt.scatter(X[:,0], X[:,1], c=clusters02, cmap='prism')  
plt.show()
plt.scatter(X[:,0], X[:,1], c=clusters03, cmap='prism')  
plt.show()
plt.scatter(X[:,0], X[:,1], c=clusters04, cmap='prism')  
plt.show()
```

# Ci-aprés la représentation des clusters pour les différents linkages utilisés en prenant comme critère le nombre de clusters désirés

```python
k=3
clusters11 = fcluster(Z1, k, criterion='maxclust')
clusters12 = fcluster(Z2, k, criterion='maxclust')
clusters13 = fcluster(Z3, k, criterion='maxclust')
clusters14 = fcluster(Z4, k, criterion='maxclust')
```

```python
plt.scatter(X[:,0], X[:,1], c=clusters11, cmap='prism')  
plt.show()
plt.scatter(X[:,0], X[:,1], c=clusters12, cmap='prism')  
plt.show()
plt.scatter(X[:,0], X[:,1], c=clusters13, cmap='prism')  
plt.show()
plt.scatter(X[:,0], X[:,1], c=clusters14, cmap='prism')  
plt.show()
```
